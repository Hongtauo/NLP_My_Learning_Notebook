```python
import math
import matplotlib.pylab as plt
#IDF算法思想

#由上TF算法得到的频次大于10的关键词以及出现的次数如下
seg_list_count=[('小区', 27), ('改造', 23), ('老旧', 22), ('朝阳区', 20), ('居民', 14), ('社会', 13), ('资本', 10)]
#假设在1000篇文档中，出现上述词的文档数为
count=[100,16,22,30,123,257,300]

idf=[]
word=[]
for i in range(len(seg_list_count)):
    word.append(seg_list_count[i][0])
    idf.append(math.log( 1000 /(1 + count[i])))

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.figure(figsize=(10, 5))
plt.plot(word,idf,'ro-')
plt.bar(word,idf)
plt.xlabel("关键词")
plt.ylabel("IDF值")
plt.title("IDF算法统计关键词频次")
plt.show()
```

```python
#根据上面的计算结果
tf= [0.037604456824512536, 0.03203342618384401, 0.03064066852367688, 0.027855153203342618, 0.019498607242339833, 0.018105849582172703, 0.013927576601671309]
idf=[2.2926347621408776, 4.074541934925921, 3.7722610630529876, 3.473768074496991, 2.0874737133771, 1.3547956940605195, 1.2006450142332614]
tf_idf=[]
for i in range(len(tf)):
    tf_idf.append(tf[i]*idf[i])

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.figure(figsize=(10, 5))
plt.plot(word,tf_idf,'ro-')
plt.bar(word,tf_idf)
plt.xlabel("关键词")
plt.ylabel("TF_IDF值")
plt.title("TF_IDF算法统计关键词频次")
plt.show()
```
